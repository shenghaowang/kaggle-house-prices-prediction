# Mising Value Imputation

```python
info_df = pd.DataFrame(df.dtypes)
info_df['MissingVal'] = df.isnull().sum()
info_df['NUnique'] = df.nunique()
info_df['Count'] = df.count()
info_df = info_df.rename(columns = {0: 'DataType'}).reset_index()
info_df = info_df.rename(columns = {'index': 'Feature'})
info_df
```

```python
missing_features = info_df[info_df['MissingVal'] > 0]
missing_features
```

```python
numeric_cols = missing_features[missing_features['DataType'] == 'float64']['Feature'].values.tolist()
for col in numeric_cols:
    print(col)
```

```python
categorical_cols = missing_features[missing_features['DataType'] == 'object']['Feature'].values.tolist()
for col in categorical_cols:
    print(col)
```
